```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([2,3,4,5])
y=np.array([4,6,8,10])

b0=0.0
b1=0.0

n=len(y)
n_iter=260
learn_rate=0.1

loss_hist=[]

for j in range(n_iter):

    y_pred=np.zeros(n)

    for i in range(n):

        y_pred[i] = b0 + b1*X[i]

        error = y_pred[i] - y[i]

        db0 = error
        db1 = error * X[i]

        b0 -= learn_rate * db0
        b1 -= learn_rate * db1

    loss = np.sum((y_pred-y)**2)/(2*n)
    loss_hist.append(loss)

    if (j %100 < 5 or j < 5):
        print(f"loss {loss:.6f}")
        print(f"b0 {b0:.6f} b1 {b1:.6f}")

plt.figure(figsize=(8,3))

plt.subplot(1,2,1)
plt.scatter(X,y,label='Data')
plt.plot(X,y_pred,'r-',label=f'Fit y={b1:.2f}X+{b0:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("SGD Linear Regression")

plt.subplot(1,2,2)
plt.plot(loss_hist[:100],'g-',label='Loss')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title("Loss Convergence")
plt.legend()

plt.tight_layout()
plt.show()
```
