```python
import numpy as np
import matplotlib.pyplot as plt

# Input values
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# OR gate outputs
y = np.array([[0],
              [1],
              [1],
              [1]])

print("OR GATE SINGLE PERCEPTRON HARD-CODED WEIGHTS")
print("-"*60)

# Hard coded weights
weights = np.array([[1.0],
                    [1.0]])

bias = np.array([[-0.5]])

print(f"Weights: W1={weights[0,0]}, W2={weights[1,0]}")
print(f"Bias: b={bias[0,0]}")
print("Equation: 1*x1 + 1*x2 - 0.5 > 0")

# Step activation
def step_function(x):
    return 1 if x > 0 else 0

# Prediction function
def perceptron_predict(X, weights, bias):
    linear = np.dot(X, weights) + bias
    predictions = np.zeros((4,1))

    for i in range(len(linear)):
        predictions[i,0] = step_function(linear[i,0])

    return predictions

print("\nPREDICTIONS")

linear_values = np.dot(X, weights) + bias
predictions = perceptron_predict(X, weights, bias)

for i in range(len(X)):
    linear_val = linear_values[i,0]
    pred = predictions[i,0]
    actual = y[i,0]

    print(f"[{X[i,0]}, {X[i,1]}]  Linear:{linear_val:.1f}  Pred:{int(pred)}  Actual:{int(actual)}")

# ---------------- GRAPH ----------------

plt.figure(figsize=(16,5))

# Decision boundary graph
plt.subplot(1,2,1)

plt.scatter(X[y.flatten()==0,0], X[y.flatten()==0,1],
            s=600, c='red', marker='o',
            edgecolors='black', label='Class 0')

plt.scatter(X[y.flatten()==1,0], X[y.flatten()==1,1],
            s=600, c='blue', marker='s',
            edgecolors='black', label='Class 1')

x1_range = np.linspace(-0.3,1.3,100)

# Decision boundary: x1 + x2 - 0.5 = 0
x2_boundary = (0.5 - x1_range)

plt.plot(x1_range, x2_boundary,
         'g', linewidth=3,
         label='Decision Boundary')

plt.xlim(-0.3,1.3)
plt.ylim(-0.3,1.3)

plt.xlabel("Input 1 (x1)")
plt.ylabel("Input 2 (x2)")
plt.title("OR Gate Decision Boundary")

plt.grid(True)
plt.legend()

# Linear combination graph
plt.subplot(1,2,2)

samples = ['[0,0]', '[0,1]', '[1,0]', '[1,1]']
linear_vals = linear_values.flatten()

plt.bar(samples, linear_vals,
        color=['red','blue','blue','blue'],
        edgecolor='black')

plt.axhline(y=0, color='black', linestyle='--')

plt.ylabel("w1x1 + w2x2 + b")
plt.title("Linear Combination Values")

plt.grid(True)

plt.tight_layout()
plt.show()

print("\nVERIFICATION RESULTS")
print("Decision boundary: x2 = 0.5 - x1")

print("\nLinear values for each input:")

for i in range(4):
    z = 1*X[i,0] + 1*X[i,1] - 0.5
    print(f"[{X[i,0]}, {X[i,1]}] : {z:.1f}")
```
