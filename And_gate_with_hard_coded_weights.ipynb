```python
import numpy as np
import matplotlib.pyplot as plt

print("AND GATE SINGLE PERCEPTRON – HARD-CODED WEIGHTS")
print("="*60)

# Input data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# Hard-coded parameters
weights = np.array([1.0,1.0])
bias = -1.5

print(f"Weights: W1={weights[0]:.1f}, W2={weights[1]:.1f}")
print(f"Bias: b={bias:.1f}")
print(f"Equation: {weights[0]:.1f}x + {weights[1]:.1f}x + {bias:.1f} > 0")

print("\nPREDICTIONS")

linear_vals = []
preds = []

for i in range(len(X)):
    linear = np.dot(X[i],weights) + bias
    pred = 1 if linear>0 else 0

    linear_vals.append(linear)
    preds.append(pred)

    print(f"{X[i]} -> Linear: {linear:.1f} -> Pred:{pred} | Actual:{y[i]}")

# ===================== PLOT =====================

plt.figure(figsize=(10,4))

# Decision boundary plot
plt.subplot(1,2,1)

for i in range(len(X)):
    if y[i]==0:
        plt.scatter(X[i,0],X[i,1],color='red',s=100)
    else:
        plt.scatter(X[i,0],X[i,1],color='blue',s=100,marker='s')

x_vals=np.linspace(-0.2,1.4,100)
y_vals=-(weights[0]*x_vals + bias)/weights[1]

plt.plot(x_vals,y_vals,'g',linewidth=2)

plt.xlim(-0.2,1.4)
plt.ylim(-0.2,1.4)

plt.xlabel("Input 1 (x1)")
plt.ylabel("Input 2 (x2)")

plt.title("Decision Boundary\n1x + 1x -1.5 = 0")

# Linear combination values plot
plt.subplot(1,2,2)

colors=['red','red','red','blue']
plt.bar(range(4),linear_vals,color=colors)

plt.axhline(0,color='black')
plt.xticks(range(4),["[0,0]","[0,1]","[1,0]","[1,1]"])

plt.title("Linear Combination Values")

plt.tight_layout()
plt.show()

# ================= VERIFICATION =================

print("\nVERIFICATION RESULTS:")
print("• Decision boundary: x = (-1x + 1.5)/1")

print("Linear values for each input:")

for i in range(len(X)):
    print(f"{X[i]}: 1*{X[i][0]} + 1*{X[i][1]} - 1 = {linear_vals[i]:.1f}")

    if linear_vals[i]>0:
        print("   Class 1 (>0)")
    else:
        print("   Class 0 (≤0)")
```
